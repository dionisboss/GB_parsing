```python

import requests
import json
import pprint
```

```python
#cоздаем класс игрок,в который будет передавать статс лист
class Player():
    def __init__(self,stats_list):
        self.stats_list = stats_list
    def max_Kills_20(self):
        max_kill_per_match = max([e['kills'] for e in self.stats_list])
        print(f'the max kills per match in the last 20 matches was: {max_kill_per_match}')
        
    def av_Duration(self):
        average_duartion = sum([e['duration'] for e in self.stats_list])/len([e['duration'] for e in self.stats_list])
        print(f'the average duration of you last 20 matches was {average_duartion}')        
     
```

# Github api

```python

import requests
import json
import pprint
```

```python
class Git_account():
    def __init__(self,git_name):
        self.git_id = git_name
        main_link = 'https://api.github.com'
        
        link_attributes = f'/users/{git_name}/repos'
        request_link = main_link + link_attributes
        try:
            response = requests.get(request_link)#,params = params)
        except:
            print('No such user')
        self.response_processed = json.loads(response.text)
    def get_Repos_name_list(self):
        repos_list = [e['name'] for e in self.response_processed]
        print(repos_list)
    def get_Full_info(self):
        try:
            pprint(self.response_processed)
        except:
            print(self.response_processed)
```

```python
while True:
    Git_user = Git_account(input('Type github user account you want to check. For default leave blank') or 'dionisboss')
    try:
        user_input = input('Type "name" to get user repos list \n or type "full" to get all repos info. \n To exit type "exit"').lower()
    except:
        continue
    if user_input == 'name':
        Git_user.get_Repos_name_list()
    elif user_input == 'full':
        Git_user.get_Full_info()
    elif user_input == 'exit':
        break
    else:
        print('Something went wrong. Try again ')
        continue 
        
        
        
```
